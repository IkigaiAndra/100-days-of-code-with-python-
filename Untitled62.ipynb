INTRODUCTION TO PANDAS

```python
# Importing Pandas
import pandas as pd

# 1. Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:")
print(df)

# 2. Basic Info about the DataFrame
print("\nBasic Information:")
print(df.info())

# 3. Descriptive statistics (summary statistics for numerical columns)
print("\nDescriptive Statistics:")
print(df.describe())

# 4. Selecting a column (e.g., 'Age')
print("\nSelect Age column:")
print(df['Age'])

# 5. Filtering data (e.g., people older than 25)
print("\nPeople older than 25:")
print(df[df['Age'] > 25])

# 6. Adding a new column ('Salary')
df['Salary'] = [50000, 60000, 45000, 70000]
print("\nDataFrame with Salary column:")
print(df)

# 7. Changing a value in the DataFrame (e.g., Alice's age)
df.loc[df['Name'] == 'Alice', 'Age'] = 25
print("\nDataFrame after changing Alice's age:")
print(df)

# 8. Dropping a column ('City')
df = df.drop(columns=['City'])
print("\nDataFrame after dropping 'City' column:")
print(df)

# 9. Sorting the DataFrame by 'Age'
print("\nDataFrame sorted by Age:")
print(df.sort_values(by='Age'))

# 10. Resetting the index after dropping rows or columns
df_reset = df.reset_index(drop=True)
print("\nDataFrame after resetting the index:")
print(df_reset)

# 11. Grouping data (grouping by Age)
print("\nGroup by Age and calculate the average Salary:")
print(df.groupby('Age')['Salary'].mean())
```

DATA CLEANING WITH PYTHON

```python
# Importing pandas
import pandas as pd
import numpy as np

# Sample DataFrame creation
data = {
    'Name': ['Alice', 'Bob', 'Charlie', None, 'David', 'Eve', 'Bob'],
    'Age': [24, 27, 22, np.nan, 32, 25, 27],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', None, 'Chicago', 'Los Angeles'],
    'Salary': [50000, 60000, 45000, 70000, 80000, np.nan, 60000]
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# 1. Handling Missing Data
# a. Filling missing values with a default value (e.g., filling NaN in 'Age' with the mean)
df['Age'] = df['Age'].fillna(df['Age'].mean())

# b. Filling missing values with a specific value (e.g., filling NaN in 'City' with 'Unknown')
df['City'] = df['City'].fillna('Unknown')

# c. Dropping rows with missing values in any column
df_dropped = df.dropna()

# 2. Removing Duplicates
# Removing duplicate rows based on all columns
df_no_duplicates = df.drop_duplicates()

# Removing duplicate entries based on specific columns (e.g., 'Name')
df_no_duplicates_name = df.drop_duplicates(subset=['Name'])

# 3. Converting Data Types
# Converting the 'Age' column to integers (if needed, after handling NaN)
df['Age'] = df['Age'].astype(int)

# 4. String Manipulations
# a. Stripping leading/trailing whitespaces from string columns
df['City'] = df['City'].str.strip()

# b. Converting text to lowercase for consistency
df['City'] = df['City'].str.lower()

# c. Replacing specific values in a column (e.g., replacing 'unknown' with 'Not Specified')
df['City'] = df['City'].replace('unknown', 'Not Specified')

# 5. Handling Outliers
# For example, let's cap the 'Salary' values that are above 75000 to 75000 (limiting outliers)
df['Salary'] = np.where(df['Salary'] > 75000, 75000, df['Salary'])

# 6. Renaming Columns
# Renaming column 'Salary' to 'Income'
df = df.rename(columns={'Salary': 'Income'})

# 7. Resetting Index after cleaning
df_reset = df.reset_index(drop=True)

# Displaying the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df)

# Displaying the DataFrame after dropping rows with NaN values
print("\nDataFrame after dropping rows with missing values:")
print(df_dropped)

# Displaying the DataFrame after removing duplicates based on 'Name' column
print("\nDataFrame after removing duplicates based on 'Name':")
print(df_no_duplicates_name)
```

DATA VISUALIZATION WITH PANDAS

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Age': [24, 27, 22, 32, 25, 29, 31],
    'Salary': [50000, 60000, 45000, 70000, 80000, 65000, 62000],
    'Department': ['HR', 'Engineering', 'Marketing', 'Engineering', 'HR', 'Marketing', 'Engineering']
}

df = pd.DataFrame(data)

# 1. Line Plot: Visualizing Salary over Age
plt.figure(figsize=(10, 6))
df.plot(x='Age', y='Salary', kind='line', marker='o', title="Salary over Age")
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()

# 2. Bar Plot: Comparing Salary across Departments
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Salary', data=df, palette='viridis')
plt.title("Salary by Department")
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()

# 3. Histogram: Distribution of Ages
plt.figure(figsize=(10, 6))
df['Age'].plot(kind='hist', bins=5, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Age Distribution")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 4. Box Plot: Salary distribution by Department
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Salary', data=df, palette='Set2')
plt.title("Salary Distribution by Department")
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()

# 5. Scatter Plot: Age vs Salary
plt.figure(figsize=(10, 6))
df.plot(kind='scatter', x='Age', y='Salary', color='green', title="Age vs Salary")
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()

# 6. Pair Plot: Pairwise relationships of all numeric columns (using Seaborn)
sns.pairplot(df[['Age', 'Salary']])
plt.show()

# 7. Heatmap: Correlation Matrix of the DataFrame
plt.figure(figsize=(8, 6))
numeric_df = df[['Age', 'Salary']]  # Select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

# 8. Pie Chart: Salary distribution by Department (percentage)
department_salary = df.groupby('Department')['Salary'].sum()
department_salary.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title("Salary Distribution by Department")
plt.ylabel('')  # Hide the y-label
plt.show()
```
