```python
import smtplib
import datetime
from email.mime.text import MIMEText

def send_reminder(to_email, subject, message, reminder_date):
    """Sends an email reminder on a specific date.
    Args:
      to_email: The recipient's email address.
      subject: The email subject.
      message: The email body.
      reminder_date: The date to send the reminder (datetime.date object).
    """
    today = datetime.date.today()
    if today == reminder_date:
        # Create and send the email
        msg = MIMEText(message)
        msg['Subject'] = subject
        msg['From'] = 'your_email@gmail.com'  # Replace with your email
        msg['To'] = to_email
        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as server:
            server.login('your_email@gmail.com', 'your_password')  # Replace with your credentials
            server.send_message(msg)
            print("Reminder email sent!")

# Get input from the user
to_email = input("Enter recipient email: ")
subject = input("Enter email subject: ")
message = input("Enter email message: ")
year = int(input("Enter reminder year (YYYY): "))
month = int(input("Enter reminder month (MM): "))
day = int(input("Enter reminder day (DD): "))
reminder_date = datetime.date(year, month, day)

# Send the reminder
send_reminder(to_email, subject, message, reminder_date)
```
