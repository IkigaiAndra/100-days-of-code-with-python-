MATH MODULE  IN PYTHON

```python
import math

# Math Constants
print("Pi:", math.pi)  # π constant
print("Euler's number:", math.e)  # e constant
print("Tau:", math.tau)  # τ constant
print("Infinity:", math.inf)  # Positive infinity
print("NaN:", math.nan)  # Not-a-Number (NaN)

# Basic Mathematical Functions
print("\nBasic Mathematical Functions")
print("Absolute value of -7:", math.fabs(-7))  # Absolute value of a number
print("Ceiling of 3.4:", math.ceil(3.4))  # Ceiling of a number (round up)
print("Floor of 3.7:", math.floor(3.7))  # Floor of a number (round down)
print("Factorial of 5:", math.factorial(5))  # Factorial of a number
print("Sum of [1, 2, 3]:", math.fsum([1, 2, 3]))  # Accurate floating-point sum

# Trigonometric Functions
print("\nTrigonometric Functions")
angle_rad = math.radians(45)  # Convert 45 degrees to radians
print("Sine of 45 degrees:", math.sin(angle_rad))
print("Cosine of 45 degrees:", math.cos(angle_rad))
print("Tangent of 45 degrees:", math.tan(angle_rad))
print("Arcsine of 0.5:", math.asin(0.5))  # Inverse sine (radians)
print("Arccosine of 0.5:", math.acos(0.5))  # Inverse cosine (radians)
print("Arctangent of 1:", math.atan(1))  # Inverse tangent (radians)
print("Arctangent of (2, 2):", math.atan2(2, 2))  # Inverse tangent of two values (radians)

# Logarithmic Functions
print("\nLogarithmic Functions")
print("Natural logarithm of 10:", math.log(10))  # Natural log (base e)
print("Logarithm base 10 of 100:", math.log10(100))  # Log base 10
print("Logarithm base 2 of 8:", math.log2(8))  # Log base 2

# Power and Exponential Functions
print("\nPower and Exponential Functions")
print("2 raised to the power 3:", math.pow(2, 3))  # x raised to the power y
print("Square root of 16:", math.sqrt(16))  # Square root of x
print("Exponential of 2:", math.exp(2))  # e^2
print("Exponential minus 1 of 2:", math.expm1(2))  # e^x - 1

# Hyperbolic Functions
print("\nHyperbolic Functions")
print("Hyperbolic sine of 1:", math.sinh(1))  # sinh(x)
print("Hyperbolic cosine of 1:", math.cosh(1))  # cosh(x)
print("Hyperbolic tangent of 1:", math.tanh(1))  # tanh(x)
print("Inverse hyperbolic sine of 1:", math.asinh(1))  # asinh(x)
print("Inverse hyperbolic cosine of 1:", math.acosh(1))  # acosh(x)
print("Inverse hyperbolic tangent of 0.5:", math.atanh(0.5))  # atanh(x)

# Angle Conversion Functions
print("\nAngle Conversion Functions")
print("45 degrees in radians:", math.radians(45))  # Convert degrees to radians
print("0.785 radians in degrees:", math.degrees(0.785))  # Convert radians to degrees

# Combinatorics Functions
print("\nCombinatorics Functions")
print("Combinations of 5 taken 3 at a time:", math.comb(5, 3))  # n choose k
print("Permutations of 5 taken 3 at a time:", math.perm(5, 3))  # nPk

# Other Functions
print("\nOther Functions")
print("Greatest common divisor of 48 and 180:", math.gcd(48, 180))  # GCD of x and y
print("Integer square root of 16:", math.isqrt(16))  # Integer square root
print("Is 1.0000001 close to 1.0?", math.isclose(1.0000001, 1.0))  # Check if two numbers are close
print("Euclidean distance between points (1, 2) and (4, 6):", math.dist((1, 2), (4, 6)))  # Euclidean distance
print("Hypotenuse of a right triangle with sides 3 and 4:", math.hypot(3, 4))  # Hypotenuse of a right triangle

# Working with Infinity and NaN
print("\nWorking with Infinity and NaN")
print("Is infinity greater than 1000?", math.inf > 1000)  # Checking infinity
print("Is NaN equal to NaN?", math.nan == math.nan)  # NaN comparison

```
