```python
# Unit Converter Project in Python

def convert_length():
    print("\nLength Conversion")
    print("1. Meters to Kilometers")
    print("2. Meters to Miles")
    print("3. Meters to Yards")
    print("4. Kilometers to Meters")
    print("5. Miles to Meters")
    print("6. Yards to Meters")

    choice = int(input("Enter the conversion option (1-6): "))
    value = float(input("Enter the value to convert: "))

    if choice == 1:
        result = value / 1000
        print(f"{value} meters = {result} kilometers")
    elif choice == 2:
        result = value * 0.000621371
        print(f"{value} meters = {result} miles")
    elif choice == 3:
        result = value * 1.09361
        print(f"{value} meters = {result} yards")
    elif choice == 4:
        result = value * 1000
        print(f"{value} kilometers = {result} meters")
    elif choice == 5:
        result = value / 0.000621371
        print(f"{value} miles = {result} meters")
    elif choice == 6:
        result = value / 1.09361
        print(f"{value} yards = {result} meters")
    else:
        print("Invalid choice!")

def convert_mass():
    print("\nMass Conversion")
    print("1. Grams to Kilograms")
    print("2. Grams to Pounds")
    print("3. Kilograms to Grams")
    print("4. Kilograms to Pounds")
    print("5. Pounds to Kilograms")
    print("6. Pounds to Grams")

    choice = int(input("Enter the conversion option (1-6): "))
    value = float(input("Enter the value to convert: "))

    if choice == 1:
        result = value / 1000
        print(f"{value} grams = {result} kilograms")
    elif choice == 2:
        result = value * 0.00220462
        print(f"{value} grams = {result} pounds")
    elif choice == 3:
        result = value * 1000
        print(f"{value} kilograms = {result} grams")
    elif choice == 4:
        result = value * 2.20462
        print(f"{value} kilograms = {result} pounds")
    elif choice == 5:
        result = value / 2.20462
        print(f"{value} pounds = {result} kilograms")
    elif choice == 6:
        result = value * 453.592
        print(f"{value} pounds = {result} grams")
    else:
        print("Invalid choice!")

def convert_temperature():
    print("\nTemperature Conversion")
    print("1. Celsius to Fahrenheit")
    print("2. Celsius to Kelvin")
    print("3. Fahrenheit to Celsius")
    print("4. Fahrenheit to Kelvin")
    print("5. Kelvin to Celsius")
    print("6. Kelvin to Fahrenheit")

    choice = int(input("Enter the conversion option (1-6): "))
    value = float(input("Enter the value to convert: "))

    if choice == 1:
        result = (value * 9/5) + 32
        print(f"{value} Celsius = {result} Fahrenheit")
    elif choice == 2:
        result = value + 273.15
        print(f"{value} Celsius = {result} Kelvin")
    elif choice == 3:
        result = (value - 32) * 5/9
        print(f"{value} Fahrenheit = {result} Celsius")
    elif choice == 4:
        result = (value - 32) * 5/9 + 273.15
        print(f"{value} Fahrenheit = {result} Kelvin")
    elif choice == 5:
        result = value - 273.15
        print(f"{value} Kelvin = {result} Celsius")
    elif choice == 6:
        result = (value - 273.15) * 9/5 + 32
        print(f"{value} Kelvin = {result} Fahrenheit")
    else:
        print("Invalid choice!")

def main():
    print("Welcome to Unit Converter!")
    while True:
        print("\nChoose the type of conversion:")
        print("1. Length")
        print("2. Mass")
        print("3. Temperature")
        print("4. Exit")

        option = int(input("Enter your choice (1-4): "))

        if option == 1:
            convert_length()
        elif option == 2:
            convert_mass()
        elif option == 3:
            convert_temperature()
        elif option == 4:
            print("Thank you for using the Unit Converter!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

# Run the main function to start the converter
if __name__ == "__main__":
    main()
```
