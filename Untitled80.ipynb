```python
import random

# List of questions, options, and the correct answers
questions = [
    {
        "question": "What is the most famous reindeer of Santa's?",
        "options": ["Rudolph", "Dasher", "Blitzen", "Comet"],
        "answer": "Rudolph"
    },
    {
        "question": "Which country started the tradition of putting up a Christmas tree?",
        "options": ["Germany", "USA", "UK", "Canada"],
        "answer": "Germany"
    },
    {
        "question": "What is the name of the red-nosed reindeer?",
        "options": ["Rudolph", "Blitzen", "Vixen", "Cupid"],
        "answer": "Rudolph"
    },
    {
        "question": "What is the name of the fictional town in 'It's a Wonderful Life'?",
        "options": ["Bedford Falls", "Whoville", "Springfield", "Gotham"],
        "answer": "Bedford Falls"
    },
    {
        "question": "In the song 'The Twelve Days of Christmas,' what did my true love give to me on the 5th day?",
        "options": ["Five golden rings", "Five silver bells", "Five candlesticks", "Five pinecones"],
        "answer": "Five golden rings"
    },
    {
        "question": "Who wrote the famous Christmas novel 'A Christmas Carol'?",
        "options": ["Charles Dickens", "J.K. Rowling", "Mark Twain", "Ernest Hemingway"],
        "answer": "Charles Dickens"
    },
    {
        "question": "What is the best-selling Christmas song of all time?",
        "options": ["Last Christmas", "Do They Know It's Christmas?", "Happy Xmas (War Is Over)", "White Christmas"],
        "answer": "White Christmas"
    },
    {
        "question": "In 'Home Alone', where are the McCallisters going on vacation when they leave Kevin behind?",
        "options": ["Paris", "London", "New York", "Rome"],
        "answer": "Paris"
    },
]

# Function to ask questions and check answers
def ask_question(question_data):
    print(f"Question: {question_data['question']}")
    print("Options:")
    for i, option in enumerate(question_data['options'], 1):
        print(f"{i}. {option}")

    # Get user's answer
    user_answer = input("Please enter the number of your answer: ")

    # Check if the answer is correct
    correct_answer = question_data['answer']
    if question_data['options'][int(user_answer) - 1] == correct_answer:
        return True
    return False

# Main trivia game function
def christmas_trivia():
    print("Welcome to the Christmas Trivia Game!")
    print("Answer the following questions to the best of your knowledge.\n")

    # Shuffle questions for a random order
    random.shuffle(questions)

    score = 0
    total_questions = len(questions)

    # Ask each question
    for i, question_data in enumerate(questions, 1):
        print(f"\nQuestion {i} of {total_questions}:")

        # Ask the question and check the answer
        if ask_question(question_data):
            print("Correct!")
            score += 1
        else:
            print(f"Wrong! The correct answer was: {question_data['answer']}")

    # Display final score
    print(f"\nYour final score is: {score}/{total_questions}")
    print("Thanks for playing!")

# Run the trivia game
if __name__ == "__main__":
    christmas_trivia()
```
